```python
#Lets convert Overtime and Attrition to integer value.

df['Attrition'] = df['Attrition'].replace(to_replace = "Yes", value = 1)
df['Attrition'] = df['Attrition'].replace(to_replace = "No", value = 0)

df['OverTime'] = df['OverTime'].replace(to_replace = "Yes", value = 1)
df['OverTime'] = df['OverTime'].replace(to_replace = "No", value = 0)

df['Over18'] = df['Over18'].replace(to_replace = "Y", value = 1)
df['Over18'] = df['Over18'].replace(to_replace = "N", value = 0)
```

```python
df.head()
```

```python
df.hist(bins = 30, figsize = (20,20), color = 'b')
```

It makes sense to drop 'EmployeeCount' ,'Standardhours' , "Over18", "Employee Count" since they do not change from one employee to the other.

```python
df.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Let's see how many employees left the company! 
left_df        = df[df['Attrition'] == 1]
stayed_df      = df[df['Attrition'] == 0]

print("Total employee in the company = ", len(df))
print("People who left the company = ", len(left_df))
print("Percentage of employees who left the company =", len(left_df)/len(df)*100.0, "%")

print("People who Stayed in the company = ", len(stayed_df))
print("Percentage of employees who stayed in the company =", len(stayed_df)/len(df)*100.0, "%")
```
